```python
import pandas as pd
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
df2=[]
scaler = MinMaxScaler()
y=[]
for path in paths:
  r=pd.read_csv(path)
  r_normalized = pd.DataFrame(scaler.fit_transform(r), columns=r.columns)
  df2.append(r_normalized.loc[:148,["x","y","z"]].values.tolist())
  y.append(list(r_normalized.loc[149,["x","y","z"]]))

df2
```

```python
len(df2[0])
```

```python
len(df2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( df2, y, test_size=0.2, random_state=42)
```

```python
len(X_test)
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense,BatchNormalization,Dropout,Bidirectional
from tensorflow.keras.losses import MeanAbsoluteError,MeanSquaredError,Huber
```

```python
from keras.regularizers import l2
```

```python
model = Sequential([
    LSTM(units=64, input_shape=(149,3), return_sequences=True, kernel_regularizer=l2(0.01)),

    LSTM(units=32, return_sequences=True, kernel_regularizer=l2(0.01)),

    LSTM(units=16,return_sequences=True, kernel_regularizer=l2(0.01)),
    LSTM(units=8, kernel_regularizer=l2(0.01)),

    Dense(units=3)
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train,validation_data=(X_test,y_test),epochs=28, batch_size=32, verbose=1)
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
mae = MeanAbsoluteError()
mse=MeanSquaredError()


loss1 = mae(y_test,pred)

loss3=mse(y_test,pred)

print("Loss%:", loss1.numpy()*100)

print("Loss%",loss3.numpy()*100)
```
